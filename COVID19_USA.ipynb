```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read csv file and put it into pandas dataframe
rawdata = pd.read_csv('us_states_covid19_daily.csv')
#print out the dataframe
print(rawdata)
```

```python
#Replace NaN value to 0.
rawdata = rawdata.replace(np.nan, 0)
```

```python
#Print out raw data without NaN value
print(rawdata)
```

```python
#Clean up unnecessary columns
cleanedData = rawdata.drop(columns=['dateChecked','hash','fips'])
```

```python
print(cleanedData)
```

```python
cleanedData.date = cleanedData.date.apply(str)
cleanedData.date = pd.to_datetime(cleanedData.date, format='%Y%m%d', errors='ignore')
cleanedData.date
```

```python
sorted_df = cleanedData.sort_values(['state','date'])
# resetting index 
sorted_df.reset_index(inplace = True) 
sorted_df.head(n=20)
```

```python
#Format figures for standard appearance.
def format_plot(fig): 

    plt.xlabel('Time')
    plt.xticks(rotation=45)
    plt.legend(loc='upper left', fontsize='large')
    

    """
    Plots states_per_plot number of best and worst faring states, ranked on the "rank_on" column.
    Labels both y axes with the y_label string provided.
    the more_is_worse boolean indicates if a larger number makes a state worse or better off.
    """
def best_and_worst(df, rank_on, y_label, states_per_plot=7, more_is_worse=True):

    most_recent_day = df.date.max()
    if more_is_worse:
        ranked = df[df['date']==most_recent_day].sort_values(by=rank_on, ascending=False)
    else:
        ranked = df[df['date']==most_recent_day].sort_values(by=rank_on, ascending=True)
    worst = ranked.head(states_per_plot).state
    best = ranked.tail(states_per_plot).state

    fig=plt.figure(figsize=(20,5))
    plt.subplot(1, 2, 1)
    for state in worst:
        state_df = df[df['state']== state].copy()
        plt.plot(state_df.date, state_df[rank_on], label=state)
        plt.title('Worst '+str(states_per_plot)+' States')
        plt.ylabel(y_label)
        format_plot(fig)
    plt.subplot(1, 2, 2)
    for state in best:
        state_df = df[df['state']== state].copy()
        plt.plot(state_df.date, state_df[rank_on], label=state)
        plt.title('Best '+str(states_per_plot)+' States')
        plt.ylabel(y_label)
        format_plot(fig)
        
    plt.show()
```

```python
#Most/least number of positive cases
best_and_worst(df=sorted_df, rank_on='positive', y_label='Number of Positive Cases')
```

```python
#Most/least number of completed tests
best_and_worst(df=sorted_df, rank_on='total', y_label='Number of Completed Tests', more_is_worse=True)
```

```python
best_and_worst(df=sorted_df, rank_on='death', y_label='Number of death due to COVID-19', more_is_worse=True)
```

```python
CA = sorted_df[sorted_df['state']=='CA']
CA
```

```python
id = []
for i in range(len(CA['date'])):
    id.append(i)
CA.insert(0, "id", id, True) 
CA
```

```python
X = CA['id'].values
y = CA['positive'].values
X = X.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(CA['id'], CA['positive'], test_size=0.25, random_state=0)
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Visualizing the Linear Regression results
def viz_linear():
    plt.scatter(X, y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('day since 03/04/2020')
    plt.ylabel('Number of positive cases')
    plt.show()
    return
viz_linear()
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)

# Visualizing the Polymonial Regression results
def viz_polymonial():
    plt.scatter(X, y, color='red')
    plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='blue')
    plt.title('Polynomial Regression of # of day vs # of positive cases confirmed')
    plt.xlabel('day since 03/04/2020')
    plt.ylabel('Number of positive cases')
    plt.show()
    return
viz_polymonial()
```
